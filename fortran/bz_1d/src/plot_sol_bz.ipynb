```python
import numpy as np

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.layouts import column

output_notebook(hide_banner=True)
```

```python
def read_sol(sol_name):
    
    file = open(sol_name, "r")
 
    # fisrt line
    line = file.readline().split()
    n = int(line[1])
    t = float(line[2])

    x=np.zeros(n)
    a=np.zeros(n)
    b=np.zeros(n)
    c=np.zeros(n)

    for i, line in enumerate(file):
        lline = line.split()
        x[i] = float(lline[0])
        a[i] = float(lline[1])
        b[i] = float(lline[2])
        c[i] = float(lline[3])
        
    return t, x, a, b, c   
```

```python
def plot_sol():

    tref, xref, aref, bref, cref = read_sol("sol_ref.dat")

    t, x, a, b, c = read_sol("sol_num.dat")
    
    dx = 4. / 4000.
    #print(np.sqrt(dx))
    #print(2/np.sqrt(4000))
    print(np.sqrt(dx)*np.linalg.norm(a-aref))

    xmin = 0.0
    xmax = 4.0

    fig_sol_a = figure(x_range=(xmin, xmax), plot_height=300, plot_width=900)
    fig_sol_b = figure(x_range=(xmin, xmax), plot_height=300, plot_width=900)
    fig_sol_c = figure(x_range=(xmin, xmax), plot_height=300, plot_width=900)
    
    fig_sol_a.x(x, a, legend="sol at t = {:4.2f}".format(t))
    fig_sol_b.x(x, b, legend="sol at t = {:4.2f}".format(t))
    fig_sol_c.x(x, c, legend="sol at t = {:4.2f}".format(t))
    
    fig_sol_a.line(xref, aref, legend="ref sol at t = {:4.2f}".format(tref), color="crimson")
    fig_sol_b.line(xref, bref, legend="ref sol at t = {:4.2f}".format(tref), color="crimson")
    fig_sol_c.line(xref, cref, legend="ref sol at t = {:4.2f}".format(tref), color="crimson")


    show(column(fig_sol_a, fig_sol_b, fig_sol_c))
    
plot_sol()
```
